```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,100)

plt.plot(x,np.sin(x),'-',c='lime')
plt.plot(x,np.cos(x))
plt.xlabel

plt.show()
```

```python
%matplotlib inline
plt.plot(x,np.cos(x))
plt.title("A cos curve")
plt.xlabel("x")
plt.ylabel("cos(y)")
# plt.show()
```
