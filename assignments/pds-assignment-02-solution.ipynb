2. Create two DataFrames: one for df1 and one for df2, each with id, name, and marks. Then use the <code>pd.concat()</code> function to combine these DataFrames.

````python
import pandas as pd

df1=pd.DataFrame(
    {
        "id":[1,2,3,4],
        "name":["Sayem","Tanim","Jilany","Selim"],
        "marks":[50,80,70,80]
    }
)
print(df1)
df2=pd.DataFrame(
    {
        "id":[5,6,7,8],
        "name":["Ramim","Tanim","Mishkat","Selim"],
        "marks":[50,80,70,80]
    }
)
concat_df=pd.concat([df1,df2],axis=1,join="outer")
print(concat_df)
````

3. Using the DataFrames from the previous task, combine them on the id column using <code>pd.merge()</code>.


````python
import matplotlib.pyplot as plt
merged_df=pd.merge(df1,df2,how="outer")
print(merged_df)

# plt.bar(merged_df["name_x"],merged_df["marks_x"])
# plt.show()
````
