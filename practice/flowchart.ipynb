```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.sankey import Sankey

# Create a basic flowchart for the scope of EVENTRY

fig, ax = plt.subplots(figsize=(10, 6))

# Define the flowchart boxes and their positions
boxes = [
    ("User Interface (UI) Creation", (0.1, 0.8)),
    ("Event Setup", (0.3, 0.6)),
    ("Marketing and Promotion", (0.5, 0.6)),
    ("Event Execution", (0.7, 0.6)),
    ("Post-Event Analytics", (0.9, 0.8)),
]

# Draw the boxes
for label, (x, y) in boxes:
    ax.text(x, y, label, ha="center", va="center", bbox=dict(boxstyle="round,pad=0.3", edgecolor="black", facecolor="lightblue"))

# Draw the arrows connecting the boxes
for i in range(len(boxes) - 1):
    ax.annotate("", xy=(boxes[i+1][1][0] - 0.07, boxes[i+1][1][1]), xytext=(boxes[i][1][0] + 0.07, boxes[i][1][1]),
                arrowprops=dict(arrowstyle="->", lw=2))

# Set the limits and remove axes
ax.set_xlim(0, 1)
ax.set_ylim(0.5, 1)
ax.axis('off')

# Title
plt.title('Scope of EVENTRY (Flowchart)', fontsize=16)

plt.show()
```
