```python
import pandas as pd
```

```python
df=pd.read_csv("./placement.csv")
df.head()
```

```python
# df.drop(labels=df.columns[0],axis=1,inplace=True)
# df.head()

print(df.shape)
df=df.iloc[:,1:]
df.head()
```

```python
x=df.iloc[:,:2]
y=df.iloc[:,2:]
y
```

```python
#creating train data(feature matrix and target vector), test data
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
x_train.head(10)
```

```python
# creating model object then fitting the train data into the model
from sklearn.linear_model import LogisticRegression

clf=LogisticRegression()

clf.fit(x_train, y_train.values.ravel())

pred=clf.predict(x_test)
print(pred)
```

```python
y_test
```

```python
#accuracy checking of prediction with test data
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```
